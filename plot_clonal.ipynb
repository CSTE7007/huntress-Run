Visualizing a tree in clonal format
===================================

This example shows how to visualize an inferred tree.


```python
#pip install scphylo-tools
```

```python
import scphylo as scp

# sphinx_gallery_thumbnail_path = "_static/thumbnails/clonal.png"
```

First, we load a binary inferred single-cell genotype data.


```python
#inferred = scp.io.read(scp.ul.get_file("scphylo.datasets/test/fp_0-fn_0-na_0.ground.CFMatrix"))
inferred = scp.io.read("Demo_test.CFMatrix")
inferred
```

```python
inferred1 = scp.io.read("Demo_reconstructed.CFMatrix")
inferred1
```

```python
tree1 = scp.ul.to_tree(inferred1)
tree1
```

```python
tree = scp.ul.to_tree(inferred)
tree
```

Next we convert the inferred genotype matrix to a tree object.


Then we can draw the tree in [clonal]{.title-ref} format i.e. mutations
at the edges and cells at at the nodes of the tree.
